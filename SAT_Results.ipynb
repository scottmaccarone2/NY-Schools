```python
from sklearn.metrics import r2_score
r2_score(y2, predict_scores)
```

**NOTE:** This is the *unadjusted* r-squared value. To obtain the adjusted r-squared value, you can compute it based on formulas/algorithms from the theory:

```python
y_hat = scores_model.predict(X2)
ss_residual = sum((y2 - y_hat)**2)
ss_total = sum((y2 - np.mean(y2))**2)
r_squared = 1 - (float(ss_residual))/ss_total
print(r_squared)
```

In the above code block, I defined `y_hat`, which is exactly the same as `predict_scores`, but I wanted to have more "statistical looking" definitions. Moreover, the r-squared value matches the `r2_score` computed previously. Now I will computed the adjusted r-squared value:

```python
adj_r2 = 1 - (1-r_squared)*(len(y1)-1)/(len(y1)-X1.shape[1]-1)
print(adj_r2)
```
